```python
#set up dependencies and make short names
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

import seaborn as sns
import seaborn.objects as so
```

```python
#import csv in the same directory as the notebook
bdf=pd.read_csv('beecrossdata.csv')
bdf

```

```python
#make a violin plot
g = sns.catplot(data=bdf, x="dose", y="crosses", hue= "dose", palette="deep", kind="violin", inner=None)
sns.swarmplot(data=bdf, x="dose", y="crosses", color="k", size=3)
plt.legend([],[], frameon=False)
ax = plt.gca()
ax.set_xticklabels(["0", ".00001", ".0001",".001",".01"])
plt.xlabel("Dose")
plt.ylabel("Number of Line Crosses")
```

```python
#make a boxplot
g = sns.catplot(
    data=bdf,
    x="crosses", y="dose",
    
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,
)
g.set(xlabel="Crosses per Hour", ylabel="")
```

```python
#import csv in the same directory as the notebook
prcsul=pd.read_csv('caffmaster.csv')
prcsul

```

```python
sns.catplot(
    data=prcsul, x="sched", y="ulcon", hue="caff",
    kind="violin",inner="stick",split="true",
)

#you need this to customize axes
ax = plt.gca()
ax.yaxis.set_major_locator(MultipleLocator(25))
ax.set_xticklabels(["30 %", "45 %", "60%"])
plt.xlabel("Sucrose Concentration")
plt.ylabel("Microliters Consumed")
#plt.ylim(0, 175)
#this gets rid of labels below zero
labels = [str(val) if val >= 0 else "" for val in ax.get_yticks()]
ax.set_yticklabels(labels)
plt.show
```

```python
sns.catplot(
    data=prcsul, x="sched", y="ulpermin", hue="caff",
    kind="violin",inner="stick",split="true",
)

#you need this to customize axes
ax = plt.gca()
ax.yaxis.set_major_locator(MultipleLocator(10)) 
ax.set_xticklabels(["30 %", "45 %", "60%"])
plt.xlabel("Sucrose Concentration")
plt.ylabel("Microliters Consumed per Minute")
#plt.ylim(0, 175)
#this gets rid of labels below zero
labels = [str(val) if val >= 0 else "" for val in ax.get_yticks()]
ax.set_yticklabels(labels)
plt.show
```

```python

dosetime=pd.read_csv('dosetime.csv')
dosetime
```

```python
# Draw a pointplot to show pulse as a function of three categorical factors
g = sns.catplot(
    data=dosetime, x="time", y="crosses", hue="dose",col="dose",
    capsize=.1, palette="deep", errorbar="se", linewidth = 2.5,
    kind="point", height=6, aspect=.75,
)
plt.xlabel("Minutes After Dose")
plt.ylabel("Crosses in the Past 30 Minutes")
g.despine(left=True)
```
